```python
"""
Oppgave a: Lese inn excel fil og lage array av kolonnenavn.
Trenger pandas for aa kunne lese excelfilen og importerer derfor denne.
Trenger også numpy pakken for aa lage arrayer.
"""

# Import av pakker
import pandas as pd
import numpy as np

# Leser fil, laster den opp samme sted som prosjektoppgaven leveres på GitHub. Kan da bruke kun filnavnet siden det er samme mappe:
data = pd.read_excel(r"support_uke_24.xlsx")

# Ser på hvordan dataen ser ut
print(data)

# Gir 217 rader (henvendelser) og 4 ulike kolonner med data

# Leser av kolonnenavn fra printutskriften. Kolonne 1 heter "Ukedag" osv.
# Lagrer kolonnenavn i en array
u_dag = np.array(data["Ukedag"])             
kl_slett = np.array(data["Klokkeslett"])
varighet = np.array(data["Varighet"])
score = np.array(data["Tilfredshet"])
```

```python
"""
b) Finn antall henvendelser per dag og visualiser dette i et stolpediagram.
Hver rad i datasettet tilsvarer en henvendelse og alle rader har en ukedag. 
Jeg kan da telle antall ukedager og bruke dette som utgangspunkt fo en visualisering.
""" 

# Sjekker datatypen til elementene:
print(type(u_dag[0]))
# Det er en string

# Leter etter alle enkelte ukedager og teller antall. Bruker if-betingelser.
# Starter med aa definere tomme variabler foer telling:

mandag = 0
tirsdag = 0
onsdag = 0
torsdag = 0
fredag = 0

# Kombinerer en for-loekke og if betingelser for aa gaa gjennom alle radene i datasettet:

for dag in u_dag:

    if dag == "Mandag":
        mandag += 1
    if dag == "Tirsdag":
        tirsdag += 1
    if dag == "Onsdag":
        onsdag += 1
    if dag == "Torsdag":
        torsdag += 1
    if dag == "Fredag":
        fredag += 1

print("Antall henvendelser per ukedag:\n mandag:",mandag,  "\n tirsdag:", tirsdag,"\n onsdag:", onsdag, "\n torsdag:", torsdag, "\n fredag:", fredag)

# Sjekker om alle rader er telt med ved aa telle totalt antall henvendelser
sum_henvendelser = mandag + tirsdag + onsdag + torsdag + fredag

print("Det er totalt ", sum_henvendelser, "antall henvendelser.")
# gir 217, samme som den initelle print av hele datasettet - har da faatt med alle.
```

```python
# Lager et stolpediagram
# Trenger da pyplot modulen fra matplotlib pakken
import matplotlib.pyplot as plt

# Definerer x og y variablene med navn på ukedager (x = ukedag) og antall henvendelser (y=henvendelser)
ukedag = ["Mandag", "Tirsdag", "Onsdag", "Torsdag", "Fredag"]
henvendelser = [mandag, tirsdag, onsdag, torsdag, fredag]

# Definerer en fargeliste for å skille stolpene tydligere
farger = ["red", "gold", "orange", "forestgreen", "blue"]

# Definerer selve stolpediaggrammet, bruker bar modulen:
plt.bar(ukedag, henvendelser, color=farger)
plt.title("Antall henvendelser per dag til Support hos Morse")
plt.xlabel("Ukedag")
plt.ylabel("Antall henvendelser")
plt.grid()
plt.show()
```

```python
"""
c) Lengste og korteste samtaletid.
Maa bruke arrayen for varighet på samtale og lete etter den laveste og hoyeste verdien. 
Bruker min() og max() funksjonene til numpypakken.
"""

samtale_kortest = np.min(varighet)
samtale_lengst = np.max(varighet)

# Deler opp samtaletidene i timer, min og sekunder ved hjelp av split() metoden for aa kunne hente ut varigheten riktig.
# Gjoer om datatypene fra string til integer for aa kunne runde av tiul hele klokkeelementer:
[time_k, min_k, sek_k] = samtale_kortest.split(":")
time_kort  = int(time_k)
min_kort = int(min_k)
sek_kort = int(sek_k)

[time_m, min_m, sek_m] = samtale_lengst.split(":")
time_lang = int(time_m)
min_lang = int(min_m)
sek_lang = int(sek_m)

print("Den korteste samtalen varte i",time_kort, "timer,", min_kort, "minutter og", sek_kort, "sekunder.")
print("Den lengste samtalen varte i",time_lang, "timer,", min_lang, "minutter og", sek_lang, "sekunder.")

```

```python
"""
d) Regne ut gjennomsnittlig samtaletid i uke 24
Varigheten er gitt i timer, minutter og sekunder som arrayer i en liste. Vil konvertere alt til sekunder for aa saa beregne gjennomsnittet.
Bruker en for loekke til aa gaa gjennom hvert element for foerst aa splitte opp timer, minutter og sekunder, for deretter aa gjoere
de om til sekunder. Til slutt summerer jeg alle sekundene og putter dem i en ny liste som kan beregnes gjennomsnitt av.
"""

# Definere en tom liste:
varighet_sek = []

# Gaa gjennom alle henvendelsene og splitte dem opp i timer, minutter og sekunder. Gjoer om til int for aa konvertere til sekunder.
# Summerer til slutt alle sekundene:
for element in varighet:
    time, min, sek = str(element).split(":")
    time_element = int(time)*60*60          # Omregning til sekunder fra time
    min_element = int(min)*60               # Omregning til sekunder fra minutt
    sek_element = int(sek)
    varighet_sek.append(time_element+ min_element + sek_element)

# Gjoer om total varighet til arrayer for aa kunne beregne gjennomsnittet med numpy funksjonen .average():
varighet_sek_arr = np.array(varighet_sek)
varighet_snitt_sek = np.average(varighet_sek_arr)

# Gjoer om tilbake til timer, minutter og sekunder. Deler på sekunder per enhet, 
# bruker int() for å runde ned til naermeste heltall:
timer_snitt = int(varighet_snitt_sek/(60*60))
min_snitt = int(varighet_snitt_sek/60)

# Trekker ut antall timer og minutter i sekunder for aa se hvor mange sekunder som gjenstaar:
sek_snitt = int(varighet_snitt_sek - timer_snitt*60*60 - min_snitt*60)

print("Den gjennomsnittlige samtalen varte i ", timer_snitt, "timer, ", min_snitt, "minutter og ", sek_snitt, "sekunder.")
```

```python
"""
e) Sektordiagram med fordeling av henvendelser i 2-timers bolker
Trenger kun aa sjekke timeelementet i "klokkeslett", bruker en for loekke for aa telle antall henvendelser i gitte timer.
Siden alle klokkeslett er oppgitt med samme timetall som start og sluttidspunkt, velger jeg aa bruke starten som foerste henvendelse, 
og opp til for det siste klokkeslettet. Dvs feks er morgen alle henvendelser fra kl 08 og tom kl 09:59:59. Formiddag starter kl 10. 
Slik blir det ikke dobbelt telling.
"""

# Definerer 2-timers bolker:

morgen = 0          # Kl 08-10
formiddag = 0       # Kl 10-12
etter_lunsj = 0     # kl 12-14  
ettermiddag = 0     # kl 14-16

# Deler opp slik at jeg får trukket ut timene, etablerer en tom liste som kan fylles:
klokkeslett = []

# Gaar gjennom alle klokkeslettene for aa hente ut timen de er registrert:
for element in kl_slett:
    time, min, sek = str(element).split(":")
    time_element = int(time)
    klokkeslett.append(time_element)

# Gaar deretter gjennom alle timene for aa splitte og telle de i de ulike bolkene:
for time in klokkeslett:

    if time == 8 or time == 9:
        morgen += 1
    if time == 10 or time == 11:
        formiddag += 1
    if time == 12 or time == 13:
        etter_lunsj += 1
    if time == 14 or time == 15:
        ettermiddag += 1


# Sektordiagram: bruker matplotlib.pyplot igjen
henvendelser_tidspunkter = ["08-10", "10-12", "12-14", "14-16"]
henvendelser_bolker = [morgen, formiddag, etter_lunsj, ettermiddag]
uthev = [0.10, 0.1, 0.1, 0.1]

plt.pie(henvendelser_bolker, labels = henvendelser_tidspunkter, colors = farger, explode = uthev, autopct ="%1.1f%%" , shadow=True)
plt.title("Fordeling av antall henvendelser gjennom tidsbolker gjennom dagen")
plt.show()
```

```python
"""
f) Beregne supportavdelingens tilfredshet.
Svar som ikke er gitt skal ikke vaere med, fyller disse med 0, bruker numpy funksjonen nan_to_num:

"""
# Gjoer om manglende tall til 0:
score_fullstendig = np.nan_to_num(score, nan=0)

# Gaar gjennom alle score og gjoer om fra float til int slik at jeg kan lete etter tallene:
for score in score_fullstendig:
    score_int = int(score)

# Lager variabler, setter de til 0:
negativ = 0
positiv = 0
passiv = 0

# Lager for loekke for aa telle antall positive, negative og passive tilbakemeldinger:

for score in score_fullstendig:
    if score >= 9:
        positiv += 1
    if score >= 7 and score <= 8:
        passiv += 1
    if score <= 6 and score> 0:   # De manglende skal ikke telle med, tar kun med fra 1
        negativ += 1

# Beregner om til prosentandelen, bruker kun de som har lagt igjen en score:
total_ant_score = positiv + negativ + passiv

andel_pos = positiv/total_ant_score*100
andel_neg = negativ/total_ant_score*100

# Beregner NPS definert som % positive - % negative, runder av til 1 desimal:
nps = round(andel_pos - andel_neg, 1)

print("Supportavdelingens NPS for uke 24 var: ", nps, "%.")

```
