
> ## Denne _notebooken_ lager et program for å sammenligne årlige kostnader for elbiler og bensinbiler 


```python
# Definerer variabler basert på opplsyninger i oppgaveteksten

km = 12000     # [Årlig kjørelengde i km]
forEl = 5000   # [Forsikringskostnad for elbil, kroner per år]
forBe = 7500   # [Forsikringskostnad for bensinbil, kroner per år]
trafAv = 8.38  # [Trafikkforsikringsavgift for begge type biler, kroner per dag]
drivEl = 0.2   # [Drivstoffbruk elbil, kWh per km]
stroemP = 2.0  # [Strømpris for lading, kroner per kWh]
drivBe = 1.0   # [Drifstoffbruk bensinbil, kroner per km]
bomAvEl = 0.1  # [Bomavgift elbil, kroner per km]
bomAvBe = 0.3  # [Bomavgift bensinbil, kroner per km]
```

```python
"""
For elbil er de variable kostnadene avhengig av strømpris, drivstoffbruk og bomavgift. Forsikring og trafikkavgift er årlige kostnader.
Drivstoffbruk og bomagift er variable kostnader for bensinbil, men også her er forsikring og traffikkavgift årlige kostnader.
Dette gir følgende formler for årlige kostnader per biltype:
"""

trafAvAAr = trafAv*365  # [Årlig kostnad for trafikkavgift, antar ikke skuddår]

aarKostEl = km*(stroemP*drivEl + bomAvEl) + forEl + trafAvAAr    # [Årlig kostnad for elbil]
aarKostBe = km*(drivBe + bomAvBe) + forBe + trafAvAAr            # [Årlig kostnad for bensinbil]

aarKostDiff = aarKostBe - aarKostEl   # [Differanse mellom årlig kostnad for elbil og bensinbil]
```

```python
print("Årlig kostnad for elbil:", int(aarKostEl), "kr")
print("Årlig kostnad for bensinbil:", int(aarKostBe), "kr")
print("Det er", int(aarKostDiff), "kr dyrere å kjøre en bensinbil enn en elbil.")
```
