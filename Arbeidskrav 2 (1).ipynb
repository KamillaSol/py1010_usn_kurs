# Oppgave 3 - fra grader til radianer

## Progam som regner om fra grader til radianer

### Formel for omregning fra grader til radianer er gitt ved:

$v_{rad} = v_{grad}*\frac{\pi}{180}$



```python

"""Her kommer selve programkoden."""

import numpy as np  # [Trenger numpy pakken for aa bruke np.pi]

v_grad = float(input("Skriv inn gradtallet:" ))     # [Formel for at bruker skal sette et gradtall]

v_rad = v_grad*np.pi/180    # [Formel som beregner om fra grader til radianr]

print(f"{v_grad} grader er det samme som {v_rad:.3} radianer.")
```

### Oppgave 5 - areal og ytre omkrets av en figur

### I denne oppgavene trengs noen matematiske formler som vist under:

### Areal av en sirkel: $\pi*r^2$
### Omrets av en sirkel: $2\pi*r$
### Areal av en trekant: $\frac{g*h}{2}$
### Omkrets av en trekant: $a+b+x$, hvor x er hypotenusen
### Pytagoras: $hypotenus^2 = katet^2 + katet^2$, eller omskrevet som: $hypotenus = \sqrt{a^2 + b^2}$, hvor a og b er katetene.


```python
import numpy as np      # [Trenger numpy for aa beregne kvadratrot]

a = float(input("Hvor lang skal diameteren og dermed den lille kateten i trekanten være? Oppgi tall i cm med 2 desimaler og . som desimaltegn: "))   # [Input funksjon for å definere stoerrelse som brukere oensker.]
b = float(input("Hvor lang skal den store kateten i trekanten være? Oppgi tall i cm med 2 desimaler og . som desimaltegn: "))   # [Input funksjon for å definere stoerelse som brukere oensker.]

# Trenger aa beregne hypotensusen i en trekant for å finne omkretsen. Figur viser at dette er en rettvinklet trekant og da gjelder Pytagoras:
x = np.sqrt(a**2+b**2)

""" Arealet av figuren bestaar av en halvsirkel og en rettvinklet trekant. Arealet av figuren vil være halvparten av arealet til sirkelen,
mens arealet av trekanten vil være gitt ved formelen over (grunnlinje*hoeyde)/2).
Omkretsen til halvsirkelen vil på samme måte være halvpraten omkretsen til en sirkel, mens omkretesen av trekanten vil være summen av den
store katet (b) og hypotenusen (x). Totalt areal og omkrets vil være summen av disse to figurene. """

areal_halvsirkel = (np.pi*a**2)/2
areal_trekant = (b*a)/2
areal_figur = areal_halvsirkel + areal_trekant

omkrets_halvsirkel = (2*np.pi*a)/2
omkrets_trekant = b + x
omkrets_figur = omkrets_halvsirkel + omkrets_trekant

print(f"Arealet av figuren, som består av diameter {a:.2f} cm, grunnlinje {b:.2f} cm og hypotenus {x:.2f} cm, er {areal_figur:.2f} cm\u00B2.")
print(f"Den ytre omkretsen av figuren, som består av diameter {a:.2f} cm, grunnlinje {b:.2f} cm og hypotenus {x:.2f} cm, er {omkrets_figur:.2f} cm.")
```

### Oppgave 6 - Funksjonsplott

Funksjon $f(x) = -x^2 - 5$
skal plottes for x i intervallet [-10, 10]



```python
import numpy as np                  # [Trenger numpy for aa lage en array til x-aksen]
import matplotlib.pyplot as plt     #[Trenger denne pakken for aa lage et plot]

x = np.linspace(-10, 10, 200)   # [En array med punkter mellom -10 til 10 til x-aksen]
y = -x**2 - 5                   # [Funksjon av x som skal plottes paa y-aksen]

plt.plot(x, y, color="green", linestyle="dotted", label="$f(x) = (-x)^2 - 5$"),     # [Plot av funksjonen med diverse visuelle justeringer.]
plt.legend()
plt.title("Plot av funksjonen: $f(x) = (-x)^2 - 5$", fontweight = "bold")
plt.xlabel("x-akse")
plt.ylabel("y-akse")
plt.grid()
plt.show()
```
